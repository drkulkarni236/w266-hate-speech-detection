```python
#imports
import numpy as np
import pandas as pd
```

```python
#Get korean train data

from google.colab import drive
drive.mount('/content/gdrive')

korean_df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks (1)/Data_Korean/train.tsv', sep = '\t'
                       )

print("Dataframe Number of (rows, columns) : {}".format(korean_df.shape))
print(korean_df.hate.value_counts())

korean_df = korean_df[['comments','hate']]
                    
korean_df.rename(columns={'comments':'text', 'hate': 'label'}, inplace = True)


pd.options.display.max_colwidth = 200
korean_df
```

```python
#Get Korean dev data
korean_dev_df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks (1)/Data_Korean/dev.tsv', sep = '\t'
                       )

print("Dataframe Number of (rows, columns) : {}".format(korean_dev_df.shape))

print(korean_dev_df.hate.value_counts())


korean_dev_df = korean_dev_df[['comments','hate']]

korean_dev_df.rename(columns={'comments':'text', 'hate': 'label'}, inplace = True)

korean_df = pd.concat([korean_df, korean_dev_df], axis = 0)

pd.options.display.max_colwidth = 200
korean_df
```

```python
# Conver labels to 0 and 1
korean_df['label'] = korean_df['label'].replace(['hate'], 1)
korean_df['label'] = korean_df['label'].replace(['offensive'], 1)
korean_df['label'] = korean_df['label'].replace(['none'], 0)
korean_df
```

```python
korean_df.label.value_counts()
```

```python
# Balancing data set
import random
random.seed(42)

korean_df = pd.concat([korean_df[korean_df.label==1].sample(korean_df.label.value_counts()[0]),
                      korean_df[korean_df.label==0]], axis = 0 )

korean_df = korean_df.sample(frac=1)

korean_df
```

```python
#Dataset is balanced
korean_df.label.value_counts()
```

```python
# Shuffle df
random.seed(42)
korean_df = korean_df.sample(frac=1).reset_index(drop=True)
korean_df
```

```python
# train dev test split 8:1:1
korean_df_train = korean_df[:int(.8*len(korean_df))]
korean_df_dev = korean_df[int(.8*len(korean_df)):int(.9*len(korean_df))]
korean_df_test = korean_df[int(.9*len(korean_df)):]
```

```python
korean_df_train.to_csv('korean_train.csv',index=False)
korean_df_train
```

```python
korean_df_dev.reset_index(drop=True, inplace=True)
korean_df_dev.to_csv('korean_dev.csv',index=False)
korean_df_dev
```

```python
korean_df_test.reset_index(drop=True, inplace=True)
korean_df_test.to_csv('korean_test.csv',index=False)
korean_df_test
```
