```python
#imports
import numpy as np
import pandas as pd
```

```python
#Get English train data from kaggle competition #https://www.kaggle.com/c/jigsaw-toxic-comment-classification-challenge

from google.colab import drive
drive.mount('/content/gdrive')

english_df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks (1)/Data_English/train_raw.csv'
                       )

print("Dataframe Number of (rows, columns) : {}".format(english_df.shape))
print(english_df.toxic.value_counts())

english_df = english_df[['comment_text','toxic']]
                    
english_df.rename(columns={'comment_text': 'text', 'toxic': 'label'}, inplace = True)


pd.options.display.max_colwidth = 200
english_df.head(10)
```

```python
# Convert labels to 0 and 1, left over from hindi, not needed for english

# hindi_df['label'] = hindi_df['label'].replace(['HOF'], 1)
# hindi_df['label'] = hindi_df['label'].replace(['NOT'], 0)
# hindi_df
```

```python
# Balancing data set
import random
random.seed(42)

english_df = pd.concat([english_df[english_df.label==0].sample(english_df.label.value_counts()[1]),
                      english_df[english_df.label==1]], axis = 0 )

english_df = english_df.sample(frac=1)

english_df
```

```python
#Dataset is balanced
english_df.label.value_counts()
```

```python
# Shuffle df
random.seed(42)
english_df = english_df.sample(frac=1).reset_index(drop=True)
english_df
```

```python
# train dev test split 8:1:1
english_df_train = english_df[:int(.8*len(english_df))]
english_df_dev = english_df[int(.8*len(english_df)):int(.9*len(english_df))]
english_df_test = english_df[int(.9*len(english_df)):]
```

```python
english_df_train.to_csv('english_train.csv',index=False)
english_df_train
```

```python
english_df_dev.reset_index(drop=True, inplace=True)
english_df_dev.to_csv('english_dev.csv',index=False)
english_df_dev
```

```python
english_df_test.reset_index(drop=True, inplace=True)
english_df_test.to_csv('english_test.csv',index=False)
english_df_test
```
