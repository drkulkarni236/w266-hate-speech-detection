```python
#imports
import numpy as np
import pandas as pd
```

```python
#Get hindi train data for 2020

from google.colab import drive
drive.mount('/content/gdrive')

hindi_df = pd.read_excel('/content/gdrive/MyDrive/Colab Notebooks (1)/Data_Hindi/Hindi_dataset_2020/hasoc_2020_hi_train.xlsx'
                       )

print("Dataframe Number of (rows, columns) : {}".format(hindi_df.shape))
print(hindi_df.task1.value_counts())

hindi_df = hindi_df[['text','task1']]
                    
hindi_df.rename(columns={'task1': 'label'}, inplace = True)


pd.options.display.max_colwidth = 200
hindi_df.head(10)
```

```python
#Get Hindi test data for 2020
hindi_test_df = pd.read_excel('/content/gdrive/MyDrive/Colab Notebooks (1)/Data_Hindi/Hindi_dataset_2020/hasoc_2020_hi_test.xlsx'
                       )

print("Dataframe Number of (rows, columns) : {}".format(hindi_test_df.shape))

print(hindi_test_df.task1.value_counts())


hindi_test_df = hindi_test_df[['text','task1']]

hindi_test_df.rename(columns={'task1': 'label'}, inplace = True)

hindi_df = pd.concat([hindi_df, hindi_test_df], axis = 0)

pd.options.display.max_colwidth = 200
hindi_df
```

```python
#Get hindi train data for 2021

from google.colab import drive
drive.mount('/content/gdrive')

hindi_df_2021 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks (1)/Data_Hindi/hi_Hasoc2021_train.csv')

print("Dataframe Number of (rows, columns) : {}".format(hindi_df_2021.shape))
print(hindi_df_2021.task_1.value_counts())


hindi_df_2021 = hindi_df_2021[['text','task_1']]

hindi_df_2021.rename(columns={'task_1': 'label'}, inplace = True)

hindi_df = pd.concat([hindi_df, hindi_df_2021], axis = 0)

print(hindi_df.label.value_counts())

pd.options.display.max_colwidth = 200
hindi_df
```

```python
# Conver labels to 0 and 1
hindi_df['label'] = hindi_df['label'].replace(['HOF'], 1)
hindi_df['label'] = hindi_df['label'].replace(['NOT'], 0)
hindi_df
```

```python
# Balancing data set
import random
random.seed(42)

hindi_df = pd.concat([hindi_df[hindi_df.label==0].sample(hindi_df.label.value_counts()[1]),
                      hindi_df[hindi_df.label==1]], axis = 0 )

hindi_df = hindi_df.sample(frac=1)

hindi_df
```

```python
#Dataset is balanced
hindi_df.label.value_counts()
```

```python
# Shuffle df
random.seed(42)
hindi_df = hindi_df.sample(frac=1).reset_index(drop=True)
hindi_df
```

```python
# train dev test split 8:1:1
hindi_df_train = hindi_df[:int(.8*len(hindi_df))]
hindi_df_dev = hindi_df[int(.8*len(hindi_df)):int(.9*len(hindi_df))]
hindi_df_test = hindi_df[int(.9*len(hindi_df)):]
```

```python
hindi_df_train.to_csv('hindi_train.csv',index=False)
hindi_df_train
```

```python
hindi_df_dev.reset_index(drop=True, inplace=True)
hindi_df_dev.to_csv('hindi_dev.csv',index=False)
hindi_df_dev
```

```python
hindi_df_test.reset_index(drop=True, inplace=True)
hindi_df_test.to_csv('hindi_test.csv',index=False)
hindi_df_test
```
